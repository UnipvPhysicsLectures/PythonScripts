## Libraries

```python
# libraries
import numpy as np
import matplotlib.pylab as plt
```

```python
n_step = 1000
x = np.linspace(-2,2,n_step)
pot = np.cos(2.0*np.pi*x)
val = np.sin(np.pi*x)**2
cond = np.cos(np.pi*x)**2
atom_x = [-1.5,-0.5,0.5,1.5]
atom_y = [-1,-1,-1,-1]
f_size = 40
```

```python
fig,ax = plt.subplots(figsize=(14,6))

# Move the left and bottom spines to x = 0 and y = 0, respectively.
ax.spines[["left", "bottom"]].set_position(("data", 0))
ax.set_axisbelow(True)

# Hide the top and right spines.
ax.spines[["top", "right"]].set_visible(False)

# Draw arrows (as black triangles: ">k"/"^k") at the end of the axes.  In each
# case, one of the coordinates (0) is a data coordinate (i.e., y = 0 or x = 0,
# respectively) and the other one (1) is an axes coordinate (i.e., at the very
# right/top of the axes).  Also, disable clipping (clip_on=False) as the marker
# actually spills out of the axes.
# ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
# ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

# change all spines
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(2)

# increase tick width
ax.tick_params(width=4)

ax.plot(x,pot,linewidth = 3.0, label= r'$V(x)$')
ax.plot(x,val,linewidth = 3.0, label= r'$|\varphi_{val}|^2$')
ax.plot(x,cond,linewidth =3.0, label= r'$|\varphi_{cond}|^2$')
ax.plot(atom_x,atom_y,'ro',markersize=20)

# ticks
ax.set_xticks([-2,-1,1,2])
ax.tick_params(labelsize=f_size-15)
ax.set_yticks([])

# labels
ax.set_xlabel('x/a',fontsize=f_size,loc='right')
# ax.set_ylabel('E',fontsize=f_size,loc='top')

# legend
ax.legend(fontsize=f_size-20,loc='upper right')

# limits
# ax.set_xlim(-8,8)

# save
fig.set_tight_layout('tight')
# fig.savefig('../figures/directBand.png',dpi=300)
```
